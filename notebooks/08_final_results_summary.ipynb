```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results_df = pd.DataFrame({
    "Model": [
        "Baseline CNN",
        "MobileNetV2",
        "EfficientNetB0",
        "MobileNetV2 + Augmentation"
    ],
    "Test Loss": [
        0.2952,
        0.3787,
        0.3915,
        0.3608
    ],
    "Test Accuracy": [
        0.9170,
        0.8736,
        0.8592,
        0.8682
    ]
})

results_df
```

```python
results_df.to_csv("../figures/final_model_results.csv", index=False)
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    data=results_df,
    x="Model",
    y="Test Accuracy"
)

plt.title("Final Test Accuracy Comparison Across Models")
plt.xlabel("Model")
plt.ylabel("Test Accuracy")
plt.ylim(0, 1)
plt.xticks(rotation=20)

plt.tight_layout()
plt.savefig("../figures/final_model_accuracy_comparison.png", dpi=300)
plt.show()
```

## Final Results Summary

The baseline CNN achieved the highest test accuracy at 91.70%, outperforming the frozen transfer learning models. MobileNetV2 achieved 87.36%, EfficientNetB0 achieved 85.92%, and MobileNetV2 with augmentation achieved 86.82%.

These results suggest that, for this dataset, a custom CNN trained directly on MRI images learned task-specific MRI features more effectively than frozen ImageNet-pretrained feature extractors. The augmentation experiment did not improve MobileNetV2 performance, producing a slightly lower test accuracy than the non-augmented MobileNetV2 model. This may indicate that the selected augmentation transformations did not provide additional useful variation, or that fine-tuning pretrained layers may be necessary for augmentation benefits to appear.
